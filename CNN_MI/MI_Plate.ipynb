```python
import numpy as np

X_loaded = np.load('X_MI_MNIST.npz')
Y_loaded = np.load('Y_MI_MNIST.npz')
```

```python
X_loaded['a'][2]
```

```python
import matplotlib.pyplot as plt

sm = plt.cm.ScalarMappable(cmap='gnuplot', norm=plt.Normalize(vmin=0, vmax=100))
sm._A = []

fig=plt.figure(figsize=(20,10))

for epoch in range(100):
    c = sm.to_rgba(epoch)
    xmvals = np.array(X_loaded['a'][epoch][0])
    ymvals = np.array(Y_loaded['a'][epoch])

    plt.plot(xmvals, ymvals, c=c, alpha=0.1, zorder=1)
    plt.scatter(xmvals, ymvals, s=20, facecolors=[c for _ in range(6)], edgecolor='none', zorder=2)

# plt.ylim([0, 3.5])
# plt.xlim([0, 14])
plt.xlabel('I(X;M)')
plt.ylabel('I(Y;M)')
plt.title('sigmoid')

cbaxes = fig.add_axes([1.0, 0.125, 0.03, 0.8]) 
plt.colorbar(sm, label='Epoch', cax=cbaxes)
plt.tight_layout()
plt.show()
```
